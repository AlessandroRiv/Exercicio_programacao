```python
import numpy as np

def fat(n):
    k=1
    for i in range(n):
        k=(i+1)*k
    return k

fat(4)
```

##  <font color='blue'>Exercício 2</font>

```python
def binomal_rev(n,p):
    draw=[]
    for i in range(n):
        draw.append(np.random.uniform(0,1))
        if draw[i]<p:
            draw[i]=1
        else:
            draw[i]=0
            
    return sum(draw)

binomal_rev(6,0.5)    
```

```python
import matplotlib.pyplot as plt

T=200
α=0.9
x_seq=[]
ϵ=[np.random.randn() for i in range(T)]

x_seq.append(0)
for i in range(T-1):
    x_seq.append(α*x_seq[i] + ϵ[i+1])
    
plt.plot(x_seq)
plt.show()
```

##  <font color='blue'>Exercício 6</font>

```python
T=200
α1=0
α2=0.8
α3=0.98
x_seq1=[]
x_seq2=[]
x_seq3=[]
ϵ=[np.random.randn() for i in range(T)]

x_seq1.append(0)
x_seq2.append(0)
x_seq3.append(0)
for i in range(T-1):
    x_seq1.append(α1*x_seq1[i] + ϵ[i+1])
    x_seq2.append(α2*x_seq2[i] + ϵ[i+1])
    x_seq3.append(α3*x_seq3[i] + ϵ[i+1])
    
plt.plot(x_seq1,'b-',label="alfa = 0")
plt.plot(x_seq2,'r-',label="alfa = 0.8")
plt.plot(x_seq3,label="alfa = 0.98")
plt.legend()
plt.show()
```

##  <font color='blue'>Exercício 4</font>

```python
def compare(seq_a,seq_b):
    count=0
    for a,b in zip(seq_a,seq_b):
        if a==b:
            count +=1
    
    return 'true' if count==len(seq_a) else 'false'

a=(1,2,3)
b=[1,4,3]
compare(a,b)
```

##  <font color='blue'>Exercício 5</font>


```python
f=lambda x: x**2

def interpol(f,dominio=(0,1),pts=1,num=0.8):
    a,b=dominio
    interval=(b-a)/(pts+1)
    local=a
    while num>local:
        local += interval
        
    u=local
    d=local-interval
    return num*((f(u)-f(d))/(u-d)) + (u*f(d)-d*f(u))/(u-d)

interpol(f,dominio=(0,10),pts=2,num=2)
```
